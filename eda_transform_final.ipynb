Verificación de outliers usando la regla del IQR (Interquartile Range)

```python
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identificar outliers
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers
```

```python
# Aplicar a las columnas numéricas
for col in df.select_dtypes(include='number').columns:
    outliers_in_column = detect_outliers_iqr(df[col])
    print(f"Número de outliers en la columna {col.upper()}: {len(outliers_in_column)}")
```

* Si el status es Check-Out, significa que el cliente estuvo en el Hotel, por lo cuál, imputaremos el año de llegada usando el mismo año de reservation_status_date para los status que sean Check-Out. Aquí tuvimos en cuenta que si el Check-Out ocurrió en enero, el año de llegada pudo haber sido el año anterior al que está en reservation_status_date. Para contemplar estos casos, usamos también la columna total_nights, que es la cantidad total de noches que el cliente estuvo hospedado en el hotel.
* Si el status es No-Show, significa que el cliente reservó pero no llegó al hotel, por lo cuál, imputaremos el año de llegada usando el mismo año de reservation_status_date para los status que sean No-Show.
* Si el status es Canceled, significa que el cliente canceló la reserva antes de la fecha de llegada, por lo cuál, imputaremos el año de llegada usando el mismo año de reservation_status_date para los status que sean Canceled.

```python
df.reservation_status.value_counts(normalize=True)*100
```

```python
# Función para imputar arrival_date_year
def impute_arrival_year(row):
    if pd.notnull(row['arrival_date_year']):
        return row['arrival_date_year']  # Mantener el valor original si no es nulo
    
    status = row['reservation_status']
    status_year = row['reservation_status_date'].year
    
    if status in ['Canceled', 'No-Show']:
        return status_year  # Usar el mismo año del reservation_status_date
    
    elif status == 'Check-Out':
        # Verificar si el check-out fue en enero
        if row['reservation_status_date'].month == 1:
            # Verificar si total_nights sugiere que la llegada fue en el año anterior
            if row['total_nights'] >= row['reservation_status_date'].day:
                return status_year - 1  # Restar un año
        return status_year  # Usar el año actual si no cumple las condiciones anteriores
    
    return pd.NaT  # En caso de que no se pueda imputar

# Aplicar la función para imputar arrival_date_year
df['arrival_date_year'] = df.apply(impute_arrival_year, axis=1)

```

* Ahora vamos a crear una columna con la fecha de llegada completa del cliente

```python
# Función para crear la fecha de llegada
def create_arrival_date(row):
    try:
        # Intenta crear la fecha con las tres columnas
        year = int(row['arrival_date_year'])
        month = int(row['arrival_date_month'])
        day = int(row['arrival_date_day_of_month'])

        # Crear y retornar la fecha si todo está bien
        return pd.to_datetime(f'{year}-{month:02d}-{day:02d}')
    
    except (ValueError, TypeError):
        # Retorna NaT (Not a Time) si hay un error en alguna conversión
        return pd.NaT

# Aplicar la función al dataframe
df['arrival_date'] = df.apply(create_arrival_date, axis=1)
```

* Teniendo la fecha de llegada vamos a imputar los nulos de la columna reservation_status_date usando la columna total_nights y arrival_date para los status que fueron check-out y no-show

```python
# Función para imputar reservation_status_date
def impute_reservation_status_date(row):
    if pd.notnull(row['reservation_status_date']):
        return row['reservation_status_date']  # Mantener el valor original si no es nulo
    
    status = row['reservation_status']
    arrival_date = row['arrival_date']
    
    if status == 'Check-Out':
        # Sumar total_nights al arrival_date para obtener el check-out date
        return arrival_date + pd.Timedelta(days=row['total_nights'])
    
    elif status == 'No-Show':
        return arrival_date  # Usar la misma fecha de llegada
    
    # Si el status es 'Canceled', mantenemos el valor nulo
    return pd.NaT

# Aplicar la función para imputar reservation_status_date
df['reservation_status_date'] = df.apply(impute_reservation_status_date, axis=1)
```

```python
def identify_linearity(dataframe, column_combinations_list):
        """
        Identifies if the relationships between pairs of variables in a DataFrame are linear or not.

        Parameters:
        -----------
        dataframe : pandas.DataFrame
            The DataFrame containing the variables to be analyzed.

        column_combinations_list : list of tuples
            A list of tuples where each tuple contains two column names from the DataFrame to be analyzed.

        Returns:
        --------
        linear_relationships : list of tuples
            A list of tuples containing the names of the columns that have a linear relationship.

        non_linear_relationships : list of tuples
            A list of tuples containing the names of the columns that do not have a linear relationship.
        """
        linear_relationships = []
        non_linear_relationships = []

        for pair in column_combinations_list: 
            # Perform the normality test
            _, p_value1 = kstest(dataframe[pair[0]], "norm")
            _, p_value2 = kstest(dataframe[pair[1]], "norm")

            if p_value1 > 0.05 and p_value2 > 0.05:
                linear_relationships.append(pair)
            else:
                non_linear_relationships.append(pair)

        return linear_relationships, non_linear_relationships


def identify_correlations(dataframe):
    """
    Identifies correlations among numeric columns in the dataframe using Pearson or Spearman methods.

    Parameters:
    -----------
    dataframe : pandas.DataFrame
        The DataFrame containing the variables to analyze.

    Returns:
    --------
    results : dict
        A dictionary containing the correlation DataFrames. The keys are 'pearson' and 'spearman'.
        If all relationships are either linear or non-linear, only one key will be present.
    """
    # Select numeric columns
    numerics = dataframe.select_dtypes(include=np.number).columns
    
    # Generate all possible combinations of numeric columns
    num_combinations = list(combinations(numerics, 2))
    
    # Identify if the relationships are linear or non-linear
    linear, non_linear = identify_linearity(dataframe, num_combinations)
    
    # Initialize the results dictionary
    results = {}

    if linear:
        # Apply Pearson correlation for linear relationships
        linear_columns = set([item for sublist in linear for item in sublist])
        df_pearson = dataframe[list(linear_columns)].corr(method="pearson")
        results['pearson'] = df_pearson

    if non_linear:
        # Apply Spearman correlation for non-linear relationships
        non_linear_columns = set([item for sublist in non_linear for item in sublist])
        df_spearman = dataframe[list(non_linear_columns)].corr(method="spearman")
        results['spearman'] = df_spearman
    
    return results


def classify_correlations(correlation_df):
    """
    Classify the correlations in the given DataFrame into weak, moderate, and strong correlations.
    
    Parameters:
    -----------
    correlation_df : pandas.DataFrame
        DataFrame containing the correlation values between pairs of variables.
    
    Returns:
    --------
    None
    """
    weak_correlations = []
    moderate_correlations = []
    strong_correlations = []

    # To avoid duplicates, use a set to register processed pairs
    processed_pairs = set()

    for row in correlation_df.index:
        for col in correlation_df.columns:
            if row != col and (col, row) not in processed_pairs:
                corr_value = correlation_df.at[row, col]
                processed_pairs.add((row, col))
                processed_pairs.add((col, row))

                if 0.1 <= abs(corr_value) < 0.3:
                    weak_correlations.append((row, col, corr_value))
                elif 0.3 <= abs(corr_value) < 0.7:
                    moderate_correlations.append((row, col, corr_value))
                elif abs(corr_value) >= 0.7:
                    strong_correlations.append((row, col, corr_value))

    # Print the results
    print("Weak Correlations:")
    for item in weak_correlations:
        print(f"Between {item[0]} and {item[1]}: {item[2]:.2f}")

    print("\nModerate Correlations:")
    for item in moderate_correlations:
        print(f"Between {item[0]} and {item[1]}: {item[2]:.2f}")

    print("\nStrong Correlations:")
    for item in strong_correlations:
        print(f"Between {item[0]} and {item[1]}: {item[2]:.2f}")

    # Return None, as we're only printing the results
    return None
    # return weak_correlations, moderate_correlations, strong_correlations
```

* Estaciones del año en Portugal:
    * Primavera: del 21 de marzo al 20 de junio. Meses: marzo, abril, mayo, y parte de junio.
    * Verano: del 21 de junio al 20 de septiembre. Meses: junio, julio, agosto, y parte de septiembre.
    * Otoño: del 21 de septiembre al 20 de diciembre. Meses: septiembre, octubre, noviembre, y parte de diciembre.
    * Invierno: del 21 de diciembre al 20 de marzo. Meses: diciembre, enero, febrero, y parte de marzo.

```python
# Extraemos el mes y el día en formato numérico para comparaciones
df['month'] = df['arrival_date'].dt.month
df['day'] = df['arrival_date'].dt.day

# Definimos las estaciones
primavera = df[((df['month'] == 3) & (df['day'] >= 21)) | (df['month'].isin([4, 5])) | ((df['month'] == 6) & (df['day'] < 21))]
verano = df[((df['month'] == 6) & (df['day'] >= 21)) | (df['month'].isin([7, 8])) | ((df['month'] == 9) & (df['day'] < 21))]
otono = df[((df['month'] == 9) & (df['day'] >= 21)) | (df['month'].isin([10, 11])) | ((df['month'] == 12) & (df['day'] < 21))]
invierno = df[((df['month'] == 12) & (df['day'] >= 21)) | (df['month'].isin([1, 2])) | ((df['month'] == 3) & (df['day'] < 21))]
```

```python
# Crear un nuevo DataFrame con las tasas de cancelación por estación y por hotel
seasonal_cancel_rates = []

# Lista de estaciones y sus datos
seasons = [primavera, verano, otono, invierno]
season_names = ['Primavera', 'Verano', 'Otoño', 'Invierno']

# Calcular la tasa de cancelación por estación y por hotel
for i, season in enumerate(seasons):
    for hotel in ['City Hotel', 'Resort Hotel']:
        # Filtrar los datos por hotel
        hotel_data = season[season['hotel'] == hotel]
        # Calcular la tasa de cancelación
        cancel_rate = hotel_data.is_canceled.mean() * 100  # Convertir a porcentaje
        seasonal_cancel_rates.append({
            'Estación': season_names[i],
            'Hotel': hotel,
            'Tasa de Cancelación (%)': cancel_rate
        })

# Convertir los datos en un DataFrame
df_cancel_rates = pd.DataFrame(seasonal_cancel_rates)

# Ordenar el DataFrame por la tasa de cancelación de mayor a menor
df_cancel_rates_sorted = df_cancel_rates.sort_values(by='Tasa de Cancelación (%)', ascending=False)

# Crear el gráfico de barras con el DataFrame ordenado
plt.figure(figsize=(10, 6))
sns.barplot(x='Estación', y='Tasa de Cancelación (%)', hue='Hotel', data=df_cancel_rates_sorted, palette='viridis')

# Agregar título y etiquetas
plt.title('Tasa de Cancelación por Estación y Hotel')
plt.ylabel('Tasa de Cancelación (%)')
plt.xlabel('Estación')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
def imprimir_tasa_cancelacion_por_adr(df, categorias):
    for categoria in categorias:
        # Filtrar el DataFrame por la categoría de ADR
        df_categoria = df[df.adr_category == categoria]
        
        # Calcular la tasa de cancelación
        if True in df_categoria.is_canceled.value_counts():
            tasa_cancelacion = df_categoria.is_canceled.value_counts(normalize=True)[True] * 100
        else:
            tasa_cancelacion = 0
        
        # Imprimir el resultado
        print(f"Tasa de cancelación de las reservas con un adr '{categoria}': {tasa_cancelacion:.2f}%")
        
# Lista de categorías de adr
categorias_adr = ['Grátis', 'Medio-bajo', 'Medio', 'Medio-alto', 'Alto', 'Bajo']

# Llamar a la función con tu DataFrame
imprimir_tasa_cancelacion_por_adr(df, categorias_adr)
```

```python
def calcular_tasa_cancelacion_por_market_segment(df):
    # Excluir los valores nulos en market_segment
    df_filtrado = df[(df.market_segment.notna()) & (df.market_segment != 'Undefined')] # Filtramos la categoria Undefined ya que solo tiene 1 registro

    # Calcular el número total de reservas y el número total de cancelaciones por market_segment
    total_reservas = df_filtrado.groupby('market_segment').size()
    cancelaciones = df_filtrado[df_filtrado['is_canceled'] == True].groupby('market_segment').size()

    # Calcular la tasa de cancelación
    tasa_cancelacion = (cancelaciones / total_reservas) * 100

    # Convertir a DataFrame para la visualización
    tasa_cancelacion_df = tasa_cancelacion.reset_index(name='Tasa de Cancelación')

    # Ordenar de mayor a menor
    tasa_cancelacion_df = tasa_cancelacion_df.sort_values(by='Tasa de Cancelación', ascending=False)

    return tasa_cancelacion_df

# Llamar a la función y mostrar el resultado
tasa_cancelacion_market_segment = calcular_tasa_cancelacion_por_market_segment(df)
tasa_cancelacion_market_segment
```

```python
def graficar_tasa_cancelacion_por_market_segment(df):
    # Calcular la tasa de cancelación por market_segment
    tasa_cancelacion_market_segment = calcular_tasa_cancelacion_por_market_segment(df)
    
    # Configuración del gráfico
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x='market_segment', 
        y='Tasa de Cancelación', 
        data=tasa_cancelacion_market_segment,
        palette='viridis'
    )
    
    # Añadir detalles al gráfico
    plt.title('Tasa de Cancelación por Segmento de Mercado', fontsize=16)
    plt.xlabel('Segmento de Mercado', fontsize=14)
    plt.ylabel('Tasa de Cancelación (%)', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    
    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()

# Llamar a la función para graficar
graficar_tasa_cancelacion_por_market_segment(df)
```
